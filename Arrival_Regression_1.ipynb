```python
import pandas as pd
df=pd.read_csv('/content/gdrive/My Drive/DSC672/DATA/Data_dharun/Merged/arrival_merged.csv')
```

```python
df
```

```python
Df_arr = df.drop(['date','ArrivalAirport','TailNumber','FlightNumber', 'ScheduledArrivalTime', 'ActualArrivalTime','OriginAirport','ScheduledElapsedTimeMinutes','TaxiIntimeMinutes','city','time','maxtempC','mintempC','sunHour','ActualElapsedTimeMinutes', 'WheelsonTime', 'moonrise', 'moonset','sunrise','sunset','CarrierCode'], axis = 1)
```

```python
arrival = Df_arr['ArrivalDelayMinutes'] > 14
print(arrival)
```

```python
Arr_new= Df_arr[arrival]
Arr_new
```

```python
Arr_new.drop(columns=Arr_new.columns[0],axis=1,inplace=True)
Arr_new
```

```python
from sklearn.preprocessing import StandardScaler as scaler
data_scaled = scaler().fit_transform(Arr_new)

data_scaled = pd.DataFrame(data_scaled)
data_scaled.columns = Arr_new.columns
```

```python
data_scaled
```

```python
X = data_scaled.drop('ArrivalDelayMinutes', axis=1)
```

```python
y = data_scaled.ArrivalDelayMinutes
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X,y,  test_size=0.30, random_state=50000)
```

```python
from sklearn.datasets import make_regression
from sklearn.ensemble import ExtraTreesRegressor
```

```python
modelER=ExtraTreesRegressor()
modelER.fit(X_train, Y_train)
```

```python
train = modelER.predict(X_train)
val = modelER.predict(X_val)
print('Training r^2: {}'.format(metrics.r2_score(Y_train, train)))
print('Validation r^2: {}'.format(metrics.r2_score(Y_val, val)))
print('RMSE training: {}'.format(np.sqrt(metrics.mean_squared_error(Y_train, train))))
print('RMSE Validation: {}'.format(np.sqrt(metrics.mean_squared_error(Y_val, val))))
print('Accuracy training: {}'.format(np.sqrt(modelER.score(X_train,Y_train))))
print('Accuracy validation: {}'.format(np.sqrt(modelER.score(X_val, Y_val))))
```

```python
gsc = GridSearchCV(
    estimator=modelER,
    param_grid={
        'n_estimators': range(5,50 ,100),
        #'max_features': range(50,401,50),
        'min_samples_leaf': range(2,10,20),
        'min_samples_split': range(2,10,20),
    },
    cv=5
)
```

```python
grid_result = gsc.fit(X_train, Y_train)
```

```python
print(grid_result.best_params_)
```
